```python
import datetime

def current_timestamp():
    now = datetime.datetime.now()
    # milliseconds precision
    return now.strftime("%Y%m%d%H%M%S") + str(now.microsecond // 1000).zfill(3)
```

```python
f = open('run_tabddpm_seq.sh', 'w')

for dataset_name in [
    'openxc-nyc-downtown-east',
    'openxc-india-new-delhi-railway-to-aiims',
    'openxc-taiwan-highwayno2-can'
] + \
[
    'car-hacking-dos-bits',
    'car-hacking-fuzzy-bits',
    'car-hacking-rpm-bits',
    'car-hacking-gear-bits'
] + \
['syncan-flag']:
    cur_time = current_timestamp()
    cmd = f'python3 driver.py --config_partition small-scale --dataset_name {dataset_name} --model_name tabddpm --cur_time {cur_time} --order_csv_by_timestamp'
    f.write(cmd + '\n')
```
